```python
import pandas as pd
import numpy as np
import re
import json
from pandas.core.frame import DataFrame
import csv
from sklearn.model_selection import train_test_split
```

```python
path = 'F:\\第三学期\\Web_Data_Integration\\WDI_TP\\datasets\\'
f = open(path + 'lyrics14.csv',encoding="utf8")
df1=pd.read_csv(f,sep=',')
df1=df1.rename(columns={'index':'id'})
print(df1.head())
print(df1.shape)
```

```python
df11=df1[['id','songName','artistName']]
print(df11.head())
print(df11.shape)
```

```python
path = 'F:\\第三学期\\Web_Data_Integration\\WDI_TP\\datasets\\'
with open(path + 'SPARQL_88.json','r') as load_f:
    df2 = json.load(load_f) 
a = df2['results']['bindings'][1] #['song']['value']
print(a)
print(type(a))
```

```python
id1 = []
songName1 = []
artistName1 = []
albumName1 = []
for i in range(len(df2['results']['bindings'])): 
    id2 = df2['results']['bindings'][i]['id']
    id1.append(id2)
    songName2 = df2['results']['bindings'][i]['song']['value']
    songName1.append(songName2)
    artistName2 = df2['results']['bindings'][i]['artistName']['value']
    artistName1.append(artistName2)
    albumName2 = df2['results']['bindings'][i]['albumName']['value']
    albumName1.append(albumName2)
df22 = {'id':id1,'songName':songName1,'artistName':artistName1}
df22=DataFrame(df22)
print(df22.head())
print(df22.shape)
```

```python
result1 = pd.merge(df11, df22, how='inner', on=['songName','artistName'])#, 'songYear'#, 'albumName'
result1
```

```python
result1.to_csv(path + 'Lyrics_SPARQ_True.csv',sep=',',index=False)
```
